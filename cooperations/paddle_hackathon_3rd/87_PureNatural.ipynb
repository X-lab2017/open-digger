# Paddle 社区数据分析
本文使用 Paddle 社区的 GitHub 日志数据对 Paddle 社区进行深度的分析，从issue、PR数量，响应周期，解决周期，响应数量，解决数量等不同的角度统计分析出结果；同时对issue和PR的评论内容进行情感分析，对正面和负面情绪进行识别与展示。最后对 Paddle 社区的现状进行总结并对未来社区的优化提出了部分建议。

## 1 issue 和PR数量统计分析（月份/年份）

### 1.1 issue 数量分析 -- 年份

```python
#2017 - 2022 各年issue数量数组
issue_countYearArr = np.array([4362 ,4979 ,4171 ,7882, 10083, 5640])
```

自 2017 - 2019年 ，paddle社区issue 的数量浮动较为平缓，2019 开始大幅上升。猜测可能因为新冠疫情爆发，大多数软件行业的工作者通过线上协作的方式展开工作，故产生大量的issue；同时在当年第一个开源代码社区建立；制定了第一个开源License -- ‘木兰宽松许可证’。开源社区的活跃自2019开始逐步升温。2021年issue的数量达到峰值，预测在未来一段时间内会稳重有升。

## 6 issue 和 PR 评论情感分析

本小节使用Senta-bilstm模型对paddle社区2016年至今的Issue和PR的评论内容进行了情感分析。
其中评论总数达276896条，正向情感占比65.11%，负向情感占比34.89%，可以看出paddle社区整体评价偏积极。

### 6.1 以年为单位统计分析结果

从下图我们可以看出，paddle社区的评论数量从2019年开始呈现显著上升的趋势，且积极评论数与消极评论数的整体趋势相差不大，可能由于2020年新冠疫情的影响，大多数开发者采用线上协作的方式，故出现大量的评论内容，2022年的评论数量仅到7月份即和2020年几乎持平，可以推测2022年全年的评论数量会达到新高。

## 7 小结

经过以上不同维度的分析可以得出以下结论：  
1）issue和PR的数量、响应数量、解决数量以及评论数量在2017-2019年变化不大，自2020开始都会出现较大幅度的增长，而响应周期与解决周期在2020年开始也会逐渐缩短，猜测可能因为新冠疫情爆发，大多数软件行业的工作者通过线上协作的方式展开工作；同时第一个开源代码社区建立；制定了第一个开源License -- ‘木兰宽松许可证’。paddle社区的活跃度自2019开始逐步升温。预测在未来社区整体活跃度将会稳中有升。 paddle社区应在活跃度提高的同时，保证各个仓库自身项目的质量。  
2）以月为单位进行统计分析可得，各年份都会受春节假期的影响，issue和PR的数量、响应数量、解决数量以及评论数量出现小幅度下降，自2月份返工后，开始回升。从3月份到6月份处于正常工作学习事件段，各统计项基本呈现平稳的趋势；7，8月份出现小幅度的上升，可能因为高校师生在这两个月具备更多自己的时间来参与到开源项目中。9月份学生返校开始新的学期任务，各统计项均有所下降。 在2021年10月份，PR数量急速下降，paddle社区需在2022年即将到来的10月份保证社区的健康度，避免出现21年急速下降的情况。  
3）经过计算各个issue的解决周期可得，issue 解决周期在 1 天内的 issue 数量最多，1 小时内以及大于 1 年的解决数量最少；超过66％的 issue 都会在 1 个月内解决，剩余的issue解决时间则超过了 1 个月。在未来，建议paddle社区issue 的解决时间更多集中在 1 个月内，并逐渐的缩减。缩减的程度可能受多方面的影响，其中包括了 issue 的难易程度，以及涉及到的技术是否被大部分人所熟知、使用。  
4）本文对issue和PR的评论进行情感分析得出，paddle社区整体评价偏积极。在各年中积极评价的数量都高于负面评价，2017年的正面评价占比最高；各个仓库的正向评论都多于负向，说明社区中大多数项目的评价较好，其中，PaddleHelix的正向情感更是达到90%以上，而LARK，PaddleX和Paddle-Lite-Demo三个项目的负向情感超出一半。paddle社区应仔细查看项目中的评论，发现存在的问题及时纠正，扩大项目的影响力，进而提高项目积极评价的比例。同时，在2021和2020年的10月份中，评论数量出现了显著的下降，有可能是因为国庆长假的影响，但仍需要paddle社区重点查看该月项目的健康状况。
